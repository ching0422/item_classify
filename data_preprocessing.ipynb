```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_excel("result_online.xlsx")
#data.shape
#data.head(50)
cate_name = data['item_category|item_category2']
print(cate_name)
```

```python
first_cate = []
#second_cate = []
for str in data['item_category|item_category2']:
    first_cate.append(str.split('|')[0])
print(first_cate)  
```

```python
cate1 = []
#cate2 = []
for i in range(0,data.shape[0]):
  if(first_cate[i] in cate1):
    continue
  else:
    cate1.append(first_cate[i])
    
print(cate1)
len(cate1)
```

```python
import csv
with open('cate.csv', 'w', newline='') as csvfile:
  writer = csv.writer(csvfile)
  writer.writerow(['category1'])
  for idx in range(0,len(cate1)):
    writer.writerow([cate1[idx]])
```
